```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
datos = pd.read_csv('/grupo1.csv')
print(datos.head())
print(datos.describe())
plt.hist(datos['Calificacion'], bins=10, color='red', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de las Calificaciones')
plt.show()
```

```python
# MODELO DE REGRESION LINEAL
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
horas_estudio = datos[['HorasEstudio']]
calificaciones = datos['Calificacion']
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
print("Ecuación de la línea de regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")
```

```python
# VISUALIZACION DE RESULTADOS Y CALCULO DE LA CORRELACION
import numpy as np
plt.scatter(datos['HorasEstudio'], datos['Calificacion'], color='green', label='Datos')
plt.plot(datos[['HorasEstudio']], modelo.predict(horas_estudio), color='black', label='Línea de regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal: Horas de Estudio vs Calificaciones')
plt.legend()
plt.show()
coef_corr, _ = pearsonr(datos['HorasEstudio'], datos['Calificacion'])
print("El coeficiente de correlación de Pearson es:", coef_corr)
```

```python
# Cargar el conjunto de datos
datos = pd.read_csv('/grupo3.csv')
print(datos.head())
print(datos.describe())
plt.hist(datos['Calificacion'], bins=10, color='blue', edgecolor='black')
plt.xlabel('Calificaciones')
plt.ylabel('Frecuencia')
plt.title('Histograma de Calificaciones')
plt.show()
```

```python
# MODELO DE REGRESION LINEAL
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

# Preparar los datos
horas_estudio = datos[['HorasEstudio']]
calificaciones = datos['Calificacion']

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)

# Obtener la pendiente e intercepto
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
print("Ecuación de la línea de regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")

```

```python
# VISUALIZACION DE RESULTADOS Y CALCULO DE LA CORRELACION
import numpy as np
plt.scatter(datos['HorasEstudio'], datos['Calificacion'], color='red', label='Datos')
plt.plot(datos[['HorasEstudio']], modelo.predict(horas_estudio), color='orange', label='Línea de regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal: Horas de Estudio vs Calificaciones')
plt.legend()
plt.show()
coef_corr, _ = pearsonr(datos['HorasEstudio'], datos['Calificacion'])
print("El coeficiente de correlación de Pearson es:", coef_corr)
```

```python
#Interpretación de Resultados
#Interpretación del coeficiente de regresión:

#El coeficiente de regresión (pendiente) representa el cambio esperado en la calificación final por cada hora adicional dedicada al estudio. Por ejemplo, si el coeficiente es 3, significa que por cada hora adicional de estudio, se espera que la calificación final aumente en 3 puntos, asumiendo que todos los demás factores se mantienen constantes.
#Interpretación del coeficiente de correlación:

#El coeficiente de correlación de Pearson mide tanto la fuerza como la dirección de la relación lineal entre las horas de estudio y las calificaciones finales. Un valor positivo cercano a 1 indica una fuerte relación directa (a más horas de estudio, mejores calificaciones), mientras que un valor cercano a -1 indicaría una relación inversa. Un valor cercano a 0 indica una relación débil o inexistente.
#Significancia estadística del modelo de regresión:

#Para evaluar si el modelo de regresión lineal es estadísticamente significativo, se puede considerar el valor p asociado al coeficiente de regresión. Un valor p menor a 0.05 indica que es muy poco probable que la relación observada sea debido al azar, lo que sugiere que el modelo es significativo y la relación entre horas de estudio y calificaciones es real.
#Confiabilidad del modelo de regresión para predicciones:

#La confiabilidad del modelo se mide mediante el coeficiente de determinación (R^2), que indica la proporción de la varianza en las calificaciones finales que puede ser explicada por las horas de estudio. Un R^2 cercano a 1 implica que el modelo tiene una alta capacidad predictiva, mientras que un R^2 bajo sugiere que hay muchos otros factores que influyen en las calificaciones finales además de las horas de estudio.
#Conclusiones Generales
#Tras analizar los resultados de los tres conjuntos de datos, se pueden extraer las siguientes conclusiones:

#Relación Lineal: En todos los grupos se encontró una relación lineal entre las horas de estudio y las calificaciones finales, lo que indica que el tiempo dedicado al estudio tiene un impacto positivo en el rendimiento académico.
#Fuerza de la Relación: La fuerza de la relación varía entre los grupos. Algunos grupos muestran una correlación más fuerte, lo que sugiere que el tiempo de estudio es un factor más determinante en sus calificaciones.
#Significancia Estadística: Los modelos de regresión lineal son estadísticamente significativos, lo que confirma que la relación observada entre las horas de estudio y las calificaciones no es producto del azar.
#Capacidad Predictiva: Aunque los modelos muestran que el tiempo de estudio influye en las calificaciones, el coeficiente de determinación (R^2) sugiere que otros factores también juegan un papel importante. Esto implica que, si bien fomentar un mayor tiempo de estudio puede mejorar las calificaciones, no debe ser el único enfoque.
#Conclusión Final
#El análisis sugiere que existe una relación positiva entre el tiempo de estudio y las calificaciones finales de los estudiantes. Sin embargo, la variabilidad en las calificaciones no puede explicarse completamente solo por el tiempo de estudio, lo que indica la presencia de otros factores influyentes. Por lo tanto, para mejorar el rendimiento académico, se debe considerar una combinación de estrategias que incluyan no solo aumentar el tiempo de estudio, sino también mejorar la calidad del estudio, proporcionar recursos adecuados, y apoyar el bienestar general de los estudiantes.
```
